```python
import numpy as np
import matplotlib.pyplot as plt
from OAM import OAMSorting as sort
```

```python
a,b,w_0 = sort.genParams(l_sort=3,lmax=20)
sorter = sort.sorter(a,b)
sorter.coords(xymax=4*w_0)
sorter.phaseElements()
phiU = sorter.phiU
```

```python
DOE = np.abs(np.exp(1j*phiU)+np.exp(1j*sorter.Y/np.max(sorter.Y)*1000))**2
DOE2 = np.abs(np.exp(1j*sorter.phiC)+np.exp(1j*sorter.U/np.max(sorter.U)*1000))**2
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(DOE2); plt.colorbar()
plt.subplot(122)
plt.imshow(sorter.phiC); plt.colorbar()
plt.show()
```
