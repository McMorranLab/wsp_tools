```python
x = np.linspace(-10,10,2**10)
y = x
X,Y = np.meshgrid(x,y)
f = np.sign(np.sin(X + 3*np.arctan2(Y,X)))
g = np.sign(np.sin(3*np.arctan2(Y,X)))
plt.imshow(g); plt.show()
```

```python
plt.plot(x,y); plt.show()
```

```python
f1 = np.fft.fftshift(np.fft.fft(np.fft.fftshift(f)))
```

```python
plt.plot(f1); plt.show()
```

```python
def f(x, a = 100, b = [1]):
    # do stuff
    f = np.zeros_like(x)
    for k in b:
        f += np.sin(k*x)*np.exp(-x**2/a)
    return(f)
```

```python
plt.plot(x,f(x,b=[1,2,3,4,5])); plt.show()
```
