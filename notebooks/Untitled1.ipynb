```python
def f(x, a = 100, b = [1]):
    # do stuff
    f = np.zeros_like(x)
    for k in b:
        f += np.sin(k*x)*np.exp(-x**2/a)
    return(f)
```

```python
plt.plot(x,f(x,b=[1,2,3,4,5])); plt.show()
```
